```python
for i in range(1, 34):
    # spacing reflects current format
    print('   DO k=1,grid%max_ts_level')
    print(f'   ts_buf(:,:) = grid%ts_ff1i{i:02}_profile(:,:,k)')
    print(f'   CALL wrf_dm_min_reals(ts_buf(:,:),grid%ts_ff1i{i:02}_profile(:,:,k),grid%ts_buf_size*grid%max_ts_locs)')
    print('   END DO')
    print('\n')
```
